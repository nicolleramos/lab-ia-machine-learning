```python
def calcular_media(notas=None):
    soma = 0
    if notas is None or len(notas) == 0: # A condicional if faz uma verificação se a lista de elementos é vazia
        return 0
    for nota in notas:
        soma += nota
    media = soma / len(notas)
    return media

notas = [7.5, 8.0, 9.5]
notas_vazia = []
print(calcular_media(notas))
print(calcular_media())
print(calcular_media(notas_vazia))
# Obs.: No meu entendimento tratar uma lista vazia também pode significar não receber a lista dentro da função
# Então fiz 3 verificações que pudessem lidar de forma segura qualquer um dos cenários
# Utilizei IA apenas para corrigir alguns erros de síntaxe já que não me lembrava de toda a estrutura de passar
# um valor direto na função.
```

**Questão 5**

Processamento de Dados com Estruturas de Repetição

```python
meses = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho',
         'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro'] # lista com todos os meses do ano

# a) Solicitar 12 valores de vendas
vendas = [] # lista vazia de vendas
for i in range(12): # a variável i irá percorrer nos 12 elementos estabelecido na função
    valor = float(input('[@]digite as vendas de ' + meses[i] + ': ')) # para cada venda inserido no mês correspondente
    vendas.append(valor) # será armazenado na lista de vendas

# b) Total, média, maior e menor venda
total = 0
for i in range(12):
    total += vendas[i] # será atribuido cada valor em vendas no total
media = total / 12 # a média é a soma de todas as vendas dividido pelo total de meses
# aqui ele determina uma condicional em que
# pra cada valor achado no índice,
# será feito uma verificação para saber se o valor do índice
# é maior que o primeiro valor determinado como maior ou menor venda.
# se for verdade, o maior/menor valor passa a ser o valor do i
maior_venda = vendas[0]
maior_mes = meses[0]
menor_venda = vendas[0]
menor_mes = meses[0]
for i in range(12):
    if vendas[i] > maior_venda:
        maior_venda = vendas[i]
        maior_mes   = meses[i]
    if vendas[i] < menor_venda:
        menor_venda = vendas[i]
        menor_mes   = meses[i]
print('\nestastíticas gerais:')
print('[*]total anual: R$', round(total, 2)) # arredonda os valores
print('[*]média mensal: R$', round(media, 2))
print('[*]maior venda: ', maior_mes, '| R$', round(maior_venda, 2))
print('[*]menor venda: ', menor_mes, '| R$', round(menor_venda, 2))

# c) Meses com vendas acima da média
print('\nmeses acima da média*:')
# irá verificar se o mês possui a venda acima da média
# percorre o índice e retorna o valor dentro daquela posição
# para cada valor que retorna true irá ser listado
for i in range(12):
    if vendas[i] > media:
        print(meses[i], '| R$', round(vendas[i], 2))

# d) Percentual de crescimento
primeiro_mes = vendas[0]
ultimo_mes   = vendas[11]
crescimento = ((ultimo_mes - primeiro_mes) / primeiro_mes) * 100

print('\n[@]crescimento*: ')
print('janeiro: R$', round(primeiro_mes, 2))
print('dezembro: R$', round(ultimo_mes, 2))
print('crescimento:', round(crescimento, 2), '%')
```

**Questão 6**

Análise Estatística com Listas e Dicionários

```python
import statistics # biblioteca com funções de cálculos

# a) Dicionário com 5 produtos e suas notas
discos = {
    'Nickelback — Silver Side Up(2001)': [8, 7, 9, 6, 8, 7, 9],
    'Metallica — ..And Justice for All(1988)': [5, 6, 7, 5, 6, 8, 5],
    'Tamia — Tamia(1997)': [9, 10, 8, 9, 7, 9, 10],
    'Mark Morrison — Return of the Mack(1996)': [4, 5, 3, 6, 4, 5, 4],
    'Return to Dust — Return to Dust(2024': [7, 6, 8, 7, 9, 6, 7]
}

# b) Função que calcula as estatísticas
def calcular_estatisticas(notas):
    media = statistics.mean(notas)
    mediana = statistics.median(notas)
    moda = statistics.mode(notas)
    desvio = statistics.stdev(notas)
    return media, mediana, moda, desvio

# c) Relatório completo por produto
def relatorio(discos):
  for produto in discos:
      notas = discos[produto]
      media, mediana, moda, desvio = calcular_estatisticas(notas)
      print(f'[*]Vinil: {produto}')
      print('notas:', notas)
      print('média:', round(media, 2))
      print('mediana:', mediana)
      print('moda:', moda)
      print('desvio padrão:', round(desvio, 2))
      print()

relatorio(discos)

# d) Produto com melhor e pior média
melhor_produto = ''
melhor_media   = 0

pior_produto   = ''
pior_media     = 999

for produto in discos:
    notas = discos[produto]
    media, mediana, moda, desvio = calcular_estatisticas(notas)

    if media > melhor_media:
        melhor_media   = media
        melhor_produto = produto

    if media < pior_media:
        pior_media   = media
        pior_produto = produto

print('[@]melhor álbum*: ', melhor_produto, '| [@]média:', round(melhor_media, 2))
print('[@]pior álbum*:  ', pior_produto,   '| [@]média:', round(pior_media, 2))

# e) Produtos com média superior a um valor
def filtrar_por_media(produtos, valor_minimo):
    produtos_filtrados = []

    for produto in produtos:
        notas = produtos[produto]
        media, mediana, moda, desvio = calcular_estatisticas(notas)
        if media > valor_minimo:
            produtos_filtrados.append(produto)
    return produtos_filtrados

valor = float(input("\n[*]digite a média mínima para filtrar os discos: "))
filtrados = filtrar_por_media(discos, valor)

print('\n[@]títulos com média superior a', valor, ':')
for p in filtrados:
    print('[*] ', p)
```

**Questão 7**
Manipulação de Matrizes - Aplicação em Imagem

a) criar matriz 5x5 com valores aleatórios


```python
import random
# [*]matriz aleatória
imagem = [] # criando a matriz linha por linha, coluna por coluna
for i in range(5):          # percorre 5 linhas
    linha = []              # cria uma linha vazia
    for j in range(5):      # percorre 5 colunas
        pixel = random.randint(0, 255)  # gera valor aleatório dentro de um intervalo definido
        linha.append(pixel)             # adiciona o pixel na linha
    imagem.append(linha)                # adiciona a linha na matriz
print(imagem) # saída: [[4, 94, 64, 132, 23], [153, 36, 220, 104, 212], [48, 172, 181, 25, 182], [58, 167, 198, 16, 102], [3, 129, 200, 103, 107]]

# b)exibir a matriz em formato de grade
# [*]exibir em grade
def matriz(imagem):
  for i in range(5):
      for j in range(5):
          print(imagem[i][j], end="\t")  # \t = tabulação para alinhar
      print()                            # quebra de linha ao fim de cada linha
matriz(imagem)
# saída:
# 68	28	129	178	238
# 70	138	81	88	114
# 39	8	106	197	243
# 41	250	74	189	205
# 129	92	55	26	80

# c)negativo da imagem (255 - valor)
# [*]inverter valores
def negativo(imagem):
  n = []
  for i in range(5):
      linha_neg = []
      for j in range(5):
          valor_invertido = 255 - imagem[i][j]   # aplicar fórmula negativo
          linha_neg.append(valor_invertido)
      n.append(linha_neg)
  for i in range(5):
    for j in range(5):
        print(n[i][j], end="\t")
    print()
negativo(imagem)
# saída
# imagem original:
# [[9, 1, 11, 73, 68], [166, 77, 182, 116, 111], [76, 147, 24, 251, 103], [147, 152, 46, 82, 221], [186, 62, 56, 197, 95]]
# valores invertidos(255-[posição[0][0]]):
# 246	254	244	182	187
# 89	178	73	139	144
# 179	108	231	4	152
# 108	103	209	173	34
# 69	193	199	58	160

# d) Threshold(limiar = 128)
# [*]aplicar limiar (threshold)
def threshold(imagem):
  t = []
  for i in range(5):
    linha_thr = []
    for j in range(5):
        if imagem[i][j] < 128: # pixel escuro
            linha_thr.append(0)
        else: # pixel claro
            linha_thr.append(255)
    t.append(linha_thr)
  for i in range(5):
    for j in range(5):
        print(t[i][j], end="\t")
    print()
threshold(imagem)
# saída:
# imagem original:
# [[29, 238, 47, 154, 80], [120, 150, 160, 232, 97], [109, 102, 164, 89, 5], [217, 209, 142, 51, 168], [229, 84, 193, 20, 224]]
# 0	255	0	255	0
# 0	255	255	255	0
# 0	0	255	0	0
# 255	255	255	0	255
# 255	0	255	0	255

# e) Histograma
def histograma(imagem):
  f1 = 0    # 0 a 63
  f2 = 0    # 64 a 127
  f3 = 0    # 128 a 191
  f4 = 0    # 192 a 255
  for i in range(5):
      for j in range(5):
          pixel = imagem[i][j]
          if pixel >= 0 and pixel <= 63:
              f1 = f1 + 1
          elif pixel >= 64 and pixel <= 127:
              f2 = f2 + 1
          elif pixel >= 128 and pixel <= 191:
              f3 = f3 + 1
          elif pixel >= 192 and pixel <= 255:
              f4 = f4 + 1
  print('[0-63]:    ', f1, 'pixels')
  print('[64-127]:  ', f2, 'pixels')
  print('[128-191]: ', f3, 'pixels')
  print('[192-255]: ', f4, 'pixels')
histograma(imagem)
# saída(exemplo):
# matriz orginal:
# [[249, 47, 104, 117, 104], [115, 45, 104, 50, 99], [195, 166, 16, 82, 207], [144, 164, 72, 240, 153], [18, 124, 243, 55, 134]]
# histograma aplicado:
# [0-63]:     6 pixels
# [64-127]:   9 pixels
# [128-191]:  5 pixels
# [192-255]:  5 pixels
```

```python
import json
# carregar dados do arquivo json
# inicializa o sistema buscando os registros
# salvos anteriormente no arquivo
def import_data():
    try:
        arquivo = open('dba_students.json', 'r')  # abre o arquivo em modo leitura
        registros = json.load(arquivo) # carrega o conteúdo do arquivo em registros
        arquivo.close() # fecha o arquivo após a leitura
        return registros # retorna os registros encontrados
    except:
        return {} # se o arquivo não existir, retorna vazio

# salvar dados no arquivo json
# persiste todos os registros ao encerrar
def salvar_dados(matriculas):
    arquivo = open('dba_students.json', 'w')  # abre o arquivo em modo escrita
    json.dump(matriculas, arquivo) # grava o dicionário no arquivo
    arquivo.close() # fecha o arquivo após a escrita

# cadastrar novo aluno
# valida todos os campos antes de salvar
def cadastrar_aluno(matriculas):
    print('\n[@] cadastrar aluno')
    codigo = input('[*] digite o RA: ') # RA é a chave do dicionário
    if codigo in matriculas: # verifica se o RA já existe
        print('[@]RA já cadastrado no sistema!')
        return
    nome_completo = input('[*] digite o nome: ')
    if nome_completo == '': # campo obrigatório
        print('[@]nome obrigatório!')
        return
    anos_vida = input('[*] digite a idade: ')
    if anos_vida == '': # campo obrigatório
        print('[@]idade obrigatória!')
        return
    graduacao = input('[*] digite o curso: ')
    if graduacao == '': # campo obrigatório
        print('[@]curso obrigatório!')
        return

    print('[*] digite as 3 notas:')
    avaliacoes = [] # lista que armazenará as 3 notas
    for i in range(3):
        pontuacao = float(input('  nota ' + str(i+1) + ': '))
        if pontuacao < 0 or pontuacao > 10: # valida o intervalo permitido
            print('[@]nota deve ser entre 0 e 10!')
            return
        avaliacoes.append(pontuacao) # adiciona a nota na lista
    matriculas[codigo] = { # cria o registro do aluno
        'nome':  nome_completo,
        'idade': anos_vida,
        'curso': graduacao,
        'notas': avaliacoes
    }

    print('[!] aluno cadastrado com sucesso')

# buscar aluno por RA
# localiza e exibe os dados de um aluno
def buscar_aluno(matriculas):
    print('\n[@]buscar Aluno')
    codigo = input('[*]digite o RA: ')
    if codigo not in matriculas: # verifica se o RA existe
        print('[@]aluno não encontrado!')
        return

    ficha = matriculas[codigo] # acessa os dados do aluno
    media_final = (ficha['notas'][0] + ficha['notas'][1] + ficha['notas'][2]) / 3

    print('\n[*]RA:     ', codigo)
    print('[*]nome:   ', ficha['nome'])
    print('[*]idade:  ', ficha['idade'])
    print('[*]curso:  ', ficha['curso'])
    print('[*]notas:  ', ficha['notas'])
    print('[*]média:  ', round(media_final, 2))

def listar_alunos(matriculas):
    print('\n[@]lista de Alunos')
    if len(matriculas) == 0:                          # verifica se há alunos cadastrados
        print('[@]nenhum aluno cadastrado!')
        return
    for codigo in matriculas:                         # percorre cada RA no dicionário
        ficha = matriculas[codigo]
        media_final = (ficha['notas'][0] + ficha['notas'][1] + ficha['notas'][2]) / 3
        print('\n[*]RA:    ', codigo)
        print('[*]nome:  ', ficha['nome'])
        print('[*]curso: ', ficha['curso'])
        print('[*]média: ', round(media_final, 2))

def atualizar_notas(matriculas):
    print('\n[@]atualizar Notas')

    codigo = input('[*] digite o RA: ')
    if codigo not in matriculas: # verifica se o RA existe
        print('[@] aluno não encontrado!')
        return

    novas_avaliacoes = [] # lista para as novas notas
    print('[*] digite as novas 3 notas:')
    for i in range(3):
        pontuacao = float(input('  Nota ' + str(i+1) + ': '))
        if pontuacao < 0 or pontuacao > 10: # valida o intervalo permitido
            print('[@]nota deve ser entre 0 e 10!')
            return
        novas_avaliacoes.append(pontuacao)

    matriculas[codigo]['notas'] = novas_avaliacoes # substitui as notas no registro
    print('[!]notas atualizadas com sucesso!')

def remover_aluno(matriculas):
    print('\n[@]remover Aluno')

    codigo = input('[*]digite o RA: ')
    if codigo not in matriculas: # verifica se o RA existe
        print('[@]aluno não encontrado!')
        return

    del matriculas[codigo] # remove o registro do dicionário
    print('[!]aluno removido com sucesso!')

def alunos_aprovados(matriculas):
    print('\n[@]alunos Aprovados (média >= 6)')

    if len(matriculas) == 0:
        print('[@]nenhum aluno cadastrado!')
        return

    for codigo in matriculas:
        ficha = matriculas[codigo]
        media_final = (ficha['notas'][0] + ficha['notas'][1] + ficha['notas'][2]) / 3
        if media_final >= 6: # filtra apenas os aprovados
            print(ficha['nome'], '| Média:', round(media_final, 2))

# soma todas as médias e divide pela
# quantidade de alunos cadastrados
def media_geral(matriculas):
    print('\n[@]média Geral da Turma')
    if len(matriculas) == 0:
        print('[!]nenhum aluno cadastrado!')
        return
    acumulador = 0 # soma de todas as médias
    total_alunos = 0 # contador de alunos

    for codigo in matriculas:
        ficha = matriculas[codigo]
        media_final = (ficha['notas'][0] + ficha['notas'][1] + ficha['notas'][2]) / 3
        acumulador = acumulador + media_final
        total_alunos = total_alunos + 1

    media_turma = acumulador / total_alunos # média de todas as médias
    print('[*]média geral da turma:', round(media_turma, 2))

# carrega os dados e mantém o sistema
# rodando até o usuário escolher sair
matriculas = import_data() # carrega os dados ao iniciar

while True:
    print('\n[@] Sistema Siga FATEC 2.0')
    print('[1] cadastrar novo aluno')
    print('[2] buscar aluno por RA')
    print('[3] listar todos os alunos')
    print('[4] atualizar notas de um aluno')
    print('[5] remover aluno')
    print('[6] alunos aprovados')
    print('[7] média geral da turma')
    print('[0] sair')

    escolha = input('\n[@] escolha uma opção: ') # lê a opção do usuário

    if escolha == '1':
        cadastrar_aluno(matriculas)
    elif escolha == '2':
        buscar_aluno(matriculas)
    elif escolha == '3':
        listar_alunos(matriculas)
    elif escolha == '4':
        atualizar_notas(matriculas)
    elif escolha == '5':
        remover_aluno(matriculas)
    elif escolha == '6':
        alunos_aprovados(matriculas)
    elif escolha == '7':
        media_geral(matriculas)
    elif escolha == '0':
        salvar_dados(matriculas) # salva os dados antes de sair
        print('[!] dados salvos. até logoOo!')
        break
    else:
        print('[@]opção inválida!')
```
